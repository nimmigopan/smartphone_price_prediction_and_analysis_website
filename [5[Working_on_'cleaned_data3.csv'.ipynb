```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('cleaned_data3.csv')
```

```python

# to make only two categories:Android,ios
df['operating_system'].value_counts()
```

```python
def os(text):
    if 'Android' in text:
        return 'Android'
    else:
        return 'iOS'
```

```python
df['operating_system'] = df['operating_system'].apply(os)
```

```python
df['operating_system'].value_counts()
```

```python

# to remove brands with only 1 phone
df['brand_name'].value_counts()
```

```python
brand_counts = df['brand_name'].value_counts()

# Get a list of brands with more than one occurrence
brands_to_keep = brand_counts[brand_counts > 1].index.tolist()

# Filter the DataFrame to keep only rows with brands in the 'brands_to_keep' list
df = df[df['brand_name'].isin(brands_to_keep)]
```

```python
df.shape
```

```python
df.info()
```

```python

# to convert 'pixels_in_width' & 'pixels_in_height' numerical columns into 1 categorical column
df['pixels_in_height'].value_counts().sort_index()
```

```python
def resolution(num):
    if num < 720:
        return 'FWVGA'
    elif 1080 > num >= 720:
        return 'HD'
    elif num == 1080:
        return 'Full HD'
    elif 1440 > num > 1080:
        return 'Full HD+'
    elif 2160 > num >= 1440:
        return 'Quad HD'
    else:
        return 'Ultra HD'
```

```python
resolution = df['pixels_in_height'].apply(resolution)
```

```python
df.insert(18,'resolution',resolution)
```

```python
df.drop(columns=['pixels_in_width','pixels_in_height','total_no_of_cameras','extended_memory_available'],inplace=True)
```
